```python
import pandas as pd

# Ensure the 'jinja2' dependency is installed
# Run this in your terminal or notebook if it's not installed:
# pip install jinja2

# Load your dataset (ensure this is done before you start creating the pivot table)
covid_df = pd.read_csv("C:/Users/hp/Downloads/COVID DATASET 1(Sheet1).csv")

# Create the pivot table based on the given values using 'max' as a string
statewise = pd.pivot_table(covid_df, values=["Confirmed", "Deaths", "Cured"], index="State/UnionTerritory", aggfunc='max')

# Calculate Recovery Rate
statewise["Recovery Rate"] = (statewise["Cured"] * 100) / statewise["Confirmed"]

# Calculate Mortality Rate
statewise["Mortality Rate"] = (statewise["Deaths"] * 100) / statewise["Confirmed"]

# Sort by 'Confirmed' in descending order
statewise = statewise.sort_values(by="Confirmed", ascending=False)

# Apply a background gradient to the DataFrame (requires matplotlib and jinja2)
statewise_styled = statewise.style.background_gradient(cmap="cubehelix")

# Display the styled DataFrame (works in Jupyter/Colab)
statewise_styled
```

```python
#code to create bar plot of top 10 countries with active covid cases///
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the columns to use while reading the CSV
columns_to_use = ['Sno', 'Date', 'Time', 'State/UnionTerritory', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed']

# Read the CSV file with the specified columns
covid_df = pd.read_csv("C:/Users/hp/Downloads/COVID DATASET 1(Sheet1).csv", usecols=columns_to_use)

# Calculate Positive Active Cases: Confirmed - (Cured + Deaths), but ensure it is not negative
covid_df['PositiveActiveCases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])

# Replace negative values with 0 (we use max to keep positive values)
covid_df['PositiveActiveCases'] = covid_df['PositiveActiveCases'].apply(lambda x: max(0, x))

# Group by 'State/UnionTerritory' and get the max positive active cases per state
top_10_positive_active_states_cases = covid_df.groupby('State/UnionTerritory')['PositiveActiveCases'].max().reset_index()

# Sort by 'PositiveActiveCases' in descending order and select the top 10
top_10_positive_active_states_cases = top_10_positive_active_states_cases.sort_values(by='PositiveActiveCases', ascending=False).head(10)

# Display the first 50 rows to verify the dataset
print(top_10_positive_active_states_cases.head(50))

# Create a bar plot for the top 10 states with the most positive active cases
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=top_10_positive_active_states_cases, x='State/UnionTerritory', y='PositiveActiveCases', linewidth=2, edgecolor='red', palette='viridis')

# Title and labels
plt.title("Top 10 States with Most Positive Active COVID-19 Cases", fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Positive Active Cases")
plt.xlabel("State/UnionTerritory")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#code to get growth trend of corona///
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the columns to use while reading the CSV
columns_to_use = ['Sno', 'Date', 'Time', 'State/UnionTerritory', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed']

# Read the CSV file with the specified columns
covid_df = pd.read_csv("C:/Users/hp/Downloads/COVID DATASET 1(Sheet1).csv", usecols=columns_to_use)

# Convert 'Date' to datetime type for easier manipulation
covid_df['Date'] = pd.to_datetime(covid_df['Date'])

# Sort by 'State/UnionTerritory' and 'Date' for proper sequential analysis
covid_df = covid_df.sort_values(by=['State/UnionTerritory', 'Date'])

# Calculate the daily increase in confirmed cases for each state
covid_df['DailyIncrease'] = covid_df.groupby('State/UnionTerritory')['Confirmed'].diff().fillna(0)

# Now we can focus on a subset of states or just analyze all states
# Let's analyze the growth trend of confirmed cases for all states over time

plt.figure(figsize=(12, 8))

# Plot the growth trend for each state (using line plot to show trend over time)
sns.lineplot(data=covid_df, x='Date', y='DailyIncrease', hue='State/UnionTerritory', lw=2, palette='tab20')

# Title and labels
plt.title("Corona Growth Trend (Daily Increase in Confirmed Cases) Statewise", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Daily Increase in Confirmed Cases")
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#code to get growth trend of corona within top 5 states/// 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the columns to use while reading the CSV
columns_to_use = ['Sno', 'Date', 'Time', 'State/UnionTerritory', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed']

# Read the CSV file with the specified columns
covid_df = pd.read_csv("C:/Users/hp/Downloads/COVID DATASET 1(Sheet1).csv", usecols=columns_to_use)

# Convert 'Date' to datetime type for easier manipulation
covid_df['Date'] = pd.to_datetime(covid_df['Date'])

# Sort by 'State/UnionTerritory' and 'Date' for proper sequential analysis
covid_df = covid_df.sort_values(by=['State/UnionTerritory', 'Date'])

# Calculate the daily increase in confirmed cases for each state
covid_df['DailyIncrease'] = covid_df.groupby('State/UnionTerritory')['Confirmed'].diff().fillna(0)

# Group by 'State/UnionTerritory' and get the sum of daily increases
statewise_growth = covid_df.groupby('State/UnionTerritory')['DailyIncrease'].sum().reset_index()

# Sort by the total daily increase in descending order and select the top 5 states
top_5_states_growth = statewise_growth.sort_values(by='DailyIncrease', ascending=False).head(5)

# Now, filter the original dataframe to include only the top 5 states
top_states_data = covid_df[covid_df['State/UnionTerritory'].isin(top_5_states_growth['State/UnionTerritory'])]

# Plot the growth trend for the top 5 states (using line plot to show trend over time)
plt.figure(figsize=(12, 8))
sns.lineplot(data=top_states_data, x='Date', y='DailyIncrease', hue='State/UnionTerritory', lw=2, palette='Set2')

# Title and labels
plt.title("Corona Growth Trend (Top 5 States with Daily Increase in Confirmed Cases)", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Daily Increase in Confirmed Cases")
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#code to create radar chart of top 10 countries with high recovery and confirmed cases///
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data for countries and their confirmed and recovery cases (comma removed for integer values)
data = {
    'Country': ['USA', 'India', 'Brazil', 'Russia', 'Mexico', 'France', 'UK', 'Italy', 'Spain', 'Turkey', 
                'Germany', 'Argentina', 'South Africa', 'Japan', 'Indonesia', 'Canada', 'Australia', 
                'South Korea', 'Colombia', 'Egypt', 'Chile', 'Poland', 'Belgium', 'Netherlands', 'Greece', 
                'Portugal', 'Ukraine', 'Israel', 'Austria', 'Switzerland', 'Romania', 'Norway', 'Hungary', 
                'Denmark', 'Czech Republic', 'Finland', 'New Zealand', 'Slovakia'],
    'Confirmed Cases': [1000000000, 700000000, 780000000, 400000000, 450000000, 350000000, 300000000, 
                        300000000, 250000000, 250000000, 250000000, 200000000, 150000000, 150000000, 
                        130000000, 170000000, 160000000, 150000000, 140000000, 130000000, 120000000, 
                        110000000, 90000000, 85000000, 70000000, 60000000, 60000000, 50000000, 45000000, 
                        42000000, 35000000, 32000000, 31000000, 29000000, 25000000, 22000000, 20000000, 
                        19000000],  # Confirmed Cases data
    'Recovery Cases': [980000000, 440000000, 450000000, 300000000, 380000000, 300000000, 280000000, 
                       250000000, 220000000, 180000000, 200000000, 150000000, 110000000, 140000000, 
                       140000000, 150000000, 140000000, 130000000, 120000000, 90000000, 80000000, 
                       80000000, 70000000, 60000000, 60000000, 50000000, 45000000, 40000000, 38000000, 
                       35000000, 25000000, 23000000, 20000000, 18000000, 15000000, 10000000, 10000000, 
                       95000000],  # Recovery Cases data
}

# Ensure both arrays are the same length
assert len(data['Country']) == len(data['Confirmed Cases']), "Arrays have different lengths!"
assert len(data['Country']) == len(data['Recovery Cases']), "Arrays have different lengths!"

# Create a DataFrame
df = pd.DataFrame(data)

# Sort the DataFrame by Confirmed Cases in descending order and select top 10
top_10_countries = df.sort_values('Confirmed Cases', ascending=False).head(10)

# Radar chart data preparation
categories = top_10_countries['Country']
confirmed_cases = top_10_countries['Confirmed Cases']
recovery_cases = top_10_countries['Recovery Cases']

# Number of categories
N = len(categories)

# Angle for each axis
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()

# Radar chart needs to "close" the loop, so we append the first value to the end of the list
confirmed_cases = np.concatenate((confirmed_cases, [confirmed_cases.iloc[0]]))
recovery_cases = np.concatenate((recovery_cases, [recovery_cases.iloc[0]]))
angles += angles[:1]

# Create the radar chart
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

# Plot the data
ax.plot(angles, confirmed_cases, color='red', linewidth=2, label='Confirmed Cases')
ax.fill(angles, confirmed_cases, color='red', alpha=0.25)

ax.plot(angles, recovery_cases, color='blue', linewidth=2, label='Recovery Cases')
ax.fill(angles, recovery_cases, color='blue', alpha=0.25)

# Set the labels for each axis (country names)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, rotation=45, ha='right', fontsize=10)

# Title and Legend
plt.title('Top 10 Countries with Highest Confirmed & Recovery Cases', fontsize=14)
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# code to create waterfall chart of top 10 countries with high mortality rate///
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data for countries and their confirmed, recovery, and death cases (comma removed for integer values)
data = {
    'Country': ['USA', 'India', 'Brazil', 'Russia', 'Mexico', 'France', 'UK', 'Italy', 'Spain', 'Turkey', 
                'Germany', 'Argentina', 'South Africa', 'Japan', 'Indonesia', 'Canada', 'Australia', 
                'South Korea', 'Colombia', 'Egypt', 'Chile', 'Poland', 'Belgium', 'Netherlands', 'Greece', 
                'Portugal', 'Ukraine', 'Israel', 'Austria', 'Switzerland', 'Romania', 'Norway', 'Hungary', 
                'Denmark', 'Czech Republic', 'Finland', 'New Zealand', 'Slovakia'],
    'Confirmed Cases': [1000000000, 700000000, 780000000, 400000000, 450000000, 350000000, 300000000, 
                        300000000, 250000000, 250000000, 250000000, 200000000, 150000000, 150000000, 
                        130000000, 170000000, 160000000, 150000000, 140000000, 130000000, 120000000, 
                        110000000, 90000000, 85000000, 70000000, 60000000, 60000000, 50000000, 45000000, 
                        42000000, 35000000, 32000000, 31000000, 29000000, 25000000, 22000000, 20000000, 
                        19000000],  # Confirmed Cases data
    'Death Cases': [20000000, 10000000, 80000000, 70000000, 50000000, 45000000, 40000000, 35000000, 
                    30000000, 25000000, 20000000, 15000000, 12000000, 13000000, 12000000, 10000000, 
                    15000000, 12000000, 13000000, 10000000, 8000000, 7500000, 6500000, 5500000, 
                    4500000, 4300000, 3500000, 2900000, 3000000, 2400000, 2300000, 2000000, 
                    1800000, 1600000, 1400000, 1200000, 1100000]  # Death Cases data
}

# Trim the data to match the shortest array length
min_length = min(len(data['Country']), len(data['Confirmed Cases']), len(data['Death Cases']))

data['Country'] = data['Country'][:min_length]
data['Confirmed Cases'] = data['Confirmed Cases'][:min_length]
data['Death Cases'] = data['Death Cases'][:min_length]

# Calculate the mortality rate (Death Cases / Confirmed Cases) * 100
data['Mortality Rate'] = (np.array(data['Death Cases']) / np.array(data['Confirmed Cases'])) * 100

# Create a DataFrame
df = pd.DataFrame(data)

# Sort the DataFrame by Mortality Rate in descending order and select top 10
top_10_countries = df.sort_values('Mortality Rate', ascending=False).head(10)

# Reset index to ensure proper access
top_10_countries.reset_index(drop=True, inplace=True)

# Combo chart data preparation
categories = top_10_countries['Country']
mortality_rate = top_10_countries['Mortality Rate']

# Waterfall chart preparation
change = mortality_rate.diff().fillna(0)  # Calculate the difference from the previous value
change.iloc[0] = mortality_rate.iloc[0]  # The first change is equal to the first value (as there's no previous value)

# Create the Waterfall chart
fig, ax = plt.subplots(figsize=(10, 6))

# Bar colors: Green for positive, Red for negative, Blue for neutral
colors = ['green' if val > 0 else 'red' if val < 0 else 'blue' for val in change]

# Plot the Waterfall bars
ax.bar(categories, change, color=colors, label='Change in Mortality Rate')

# Add the starting point for the waterfall
ax.bar(categories, mortality_rate, color='blue', alpha=0.3, label='Mortality Rate')

# Set labels and title
ax.set_xlabel('Country', fontsize=12)
ax.set_ylabel('Mortality Rate (%)', fontsize=12)
plt.title('Waterfall Chart for Mortality Rate in Top 10 Countries', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)

# Add Legends
ax.legend(loc='upper left')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
